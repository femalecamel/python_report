### range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
> range(stop)

> range(start, stop, [step])

frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

```python
def frange(*arg) :
    if len(arg) == 1 :
        return map(lambda x: x/10.0, range(0, int(arg[0]*10), 1))
    elif len(arg) == 2 :
        return map(lambda x: x/10.0, range(int(arg[0]*10), int(arg[1]*10), 1))
    elif len(arg) == 3 :
        return map(lambda x: x/10.0, range(int(arg[0]*10), int(arg[1]*10), int(arg[2]*10)))
    else :
        return 'Wrong Input'

print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
print frange(3,5,6,7,8)
```

새로운 함수 frange를 정의할 때 반복적인 동작과 범위 설정을 해야하므로 map, lambda, range를 사용하였다. 또한 start, stop, step의 입력 여부에 따라 함수의 케이스가 달라진다. range와 달리 새로운 함수인 frange는 이를 일일히 선언해줘야하므로 가변인수의 길이에 따라 케이스를 분류한다.

range에서 받는 파라미터는 start, stop, step까지 최대 3개이다. 만약 파라미터가 3개를 초과한다면 잘못된 Input이라고 알려준다.

가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

```python
def sum(*arg) :
    result = 0
    for num in arg :
        result += num
    return result

print sum()
print sum(1, 2)
print sum(1, 2, 3, 4, 5)
print sum(1, 5, 7, 2, -10)
```

숫자의 합(결과)를 뜻하는 result를 변수로 선언하고 반복문을 이용해 합을 구한다. 이 때 함수 sum에서는 가변인수를 받도록하였다.

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n) :
    if n == 0 :
        return 1
    else :
        return n * myfact(n-1)
    
print myfact(5)
print
print myfact(50)
print
print myfact(100)
#print myfact(1000)
#print myfact(5000)
```

재귀함수를 이용해 팩토리얼을 계산하였다. 파라미터를 1000이상으로 주었더니 maximum recursion depth exceeded이라는 메세지와 함께 에러가 나와서 작은수를 넣었다.

```python
minor = range(3, 10000, 2) # 홀수 리스트

j = 3 # minor[3] = 9 부터 시작

while (j < len(minor)) : # 소수 검사
    i = 0
    while(i < j) :
        if minor[i] == 0 : # 0으로 된 수로 나누었을 때 처리법, pass
            i += 1
            continue
        if minor[j] % minor[i] == 0 :
            minor[j] = 0
        i += 1
    j += 1

#minor의 원소 합으로 소수의 합을 구함
print reduce(lambda x, y: x + y, minor) + 2
```

1부터 n까지의 자연수를 차례로 더하여 구해진 값을 삼각수라고 합니다.
예를 들어 7번째 삼각수는 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28이 됩니다.
이런 식으로 삼각수를 구해 나가면 다음과 같습니다.

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...
이 삼각수들의 약수를 구해봅시다.

- 1: 1
- 3: 1, 3
- 6: 1, 2, 3, 6
- 10: 1, 2, 5, 10
- 15: 1, 3, 5, 15
- 21: 1, 3, 7, 21
- 28: 1, 2, 4, 7, 14, 28

위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.

그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?

```python
def numberOf(num) : # 삼각수의 약수 갯수 반환
    count = 0
    for x in range(1, num+1) :
        if num%x == 0 : # 약수
            count += 1
    return count

num = 1
n = 0
while ( numberOf(num) <= 10 ) : # 500개 이상은 시간이 오래걸려서 10으로 수정
    num = (n*(n+1))/2
    n += 1
print num
```

삼각수를 구하는 공식은 (n*(n+1))/2이다. 반복문을 활용해 계산을 하고 약수의 갯수는 함수를 통해 받는다. 이 때 x로 나누어떨어지면 약수이므로 count가 올라간다.

# Euler 15 ---------------------------------------------
2 × 2 격자의 왼쪽 위 모서리에서 출발하여 오른쪽 아래 모서리까지 도달하는 길은 모두 6가지가 있습니다 (거슬러 가지는 않기로 합니다).
그러면 20 × 20 격자에는 모두 몇 개의 경로가 있습니까?

```python
import math

r1 = math.factorial(40) # 가로20개, 세로20개를 일렬로 나열
r2 = math.factorial(20) * math.factorial(20) # 경우의 수 고려

print r1/r2
```

순열조합의 관점에서 접근하면 해결하기가 수월하다. 가로, 세로로 이동하면서 경로를 완성하는 것이므로 각각의 가로, 세로를 일렬로 나열했다고 할 수 있다. 이 때 모든 경우의 수는 (a+b)! / a! / b!이다. 따라서 분모와 분자를 따로 내장함수을 이용하여 값을 받고 나누어서 결과를 구한다.
