키보드로 정수값을 입력 받고 그 값이 양수인지 음수인지를 비트 연산자를 이용하여 판단하는 프로그램을 작성하시오.

- [참고] 가장 왼쪽에 있는 비트가 0이면 양수이며, 1이면 음수이다.
- [참고] 수업보충자료 (https://github.com/bluebibi/python-e-learning/blob/master/supplement.ipynb) 를 참고하시오.

```python
def findNegative(n):
    shift = n.bit_length()/4 + 1
    shift = 0x8 << shift*4

    if ( ( n & shift ) == shift ):
        print u'입력한 %d은 음수입니다.'%(n)
    else:
        print u'입력한 %d은 양수입니다.'%(n)
        
num = input(u'정수를 입력하세요.:')
findNegative(num)
```

입력받은 정수의 길이를 이용하여 음수판단 값을 만들어 판단한다. 논리연산자 and를 통해 입력한 값의 최상위 비트가 0, 1 중 무엇인지 판단하는데 이 때 0이면 양수, 1이면 음수이다.

경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
- 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
- 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
- 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
> - div_path('/usr/local/bin/python')
> - ['/usr/local/bin', 'python']
> - div_path('/home/chulsoo/test.txt')
> - ['/home/chulsoo', 'test.txt']

- [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.

```python
def div_path(s):
    result = []
    l = s.split('/')
    result.append(l.pop())
    result.insert(0,'/'.join(l))
    return result

st = raw_input(u'경로에 해당하는 문자열을 입력하시오. :')
l = div_path(st)
print st
print l
```

입력한 경로 문자열을 '/'으로 나누어 리스트에 저장한다. 파일명은 경로 중 가장 마지막에 있으므로 맨 마지막 값을 결과 리스트에 넣어준다. 그 다음 나머지 값을 '/'와 함께 붙여서 출력하였다.

두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
- 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
- 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
- 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
- 반환하는 리스트는 정렬이 되어 있어야 한다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
> - list_union([1, 2, 3], [1, 2, 4])
> - [1, 2, 3, 4]
> - list_union([-10, -5, 0, -1], [100, 9, 0, 9])
> - [-10, -5, -1, 0, 9, 100]
> - list_union([0, 1, 2], [0, 1, 2])
> - [0, 1, 2]

- [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
- [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista, listb):
    for y in listb :
        if not ( y in lista ):
            lista.append(y)
    return sorted(lista)
print list_union([1, 2, 3], [1, 2, 4])
print list_union([-10, -5, 0, -1], [100, 9, 0, 9])
print list_union([0, 1, 2], [0, 1, 2])
```

list_union 함수를 통해 두 개의 리스트를 받아 리스트의 원소들을 검사한다. 이 때 두 번째 리스트의 원소 중 첫 번째 리스트에 없는 것이라면 첫 번재 리스트에 삽입한 후 sorted를 통해 정렬한다.

```python
def getLCM( a, b ): ## 최소공배수
    result = 1
  
    pA = GetPrimary(a)
    pB = GetPrimary(b)
    
    for x in list( set(pA) | set(pB) ):
        if (pA.count(x) > pB.count(x)):
            result *= x ** pA.count(x)
        else:
            result *= x ** pB.count(x)
    
    return result

def PrimaryN( n ): ## 소수   
    if n is 2:
        return True
    
    elif n > 2:
        for y in xrange (2, n):
            if ( n%y is 0 ):
                break
        else:
            return True
    return False

def GetPrimary( n ):  ## 소인수 삽입
    result = []

    for z in xrange(2, n) :
        if n % z == 0 :
            if PrimaryN( z ):
                result += [z]
                n /= z
                result += GetPrimary( n )
                break
    else :
        if PrimaryN( n ) :
            result += [n]
            
    return result 

def getAllNumDivide( n ):
    x = 1
    y = 2
    while ( not( y == n + 1 ) ):
        x = getLCM( x, y )
        y += 1
    return x

print "1 ~ 10까지의 어떤 수로도 나누어 떨어지는 가장 작은 수는 %d" % (getAllNumDivide(10))
print "1 ~ 20까지의 어떤 수로도 나누어 떨어지는 가장 작은 수는 %d" % (getAllNumDivide(20))
```

## Project Euler 6
1부터 10까지 자연수를 각각 제곱해 더하면 다음과 같습니다 (제곱의 합).

12 + 22 + ... + 102 = 385
1부터 10을 먼저 더한 다음에 그 결과를 제곱하면 다음과 같습니다 (합의 제곱).

(1 + 2 + ... + 10)2 = 552 = 3025
따라서 1부터 10까지 자연수에 대해 "합의 제곱"과 "제곱의 합" 의 차이는 3025 - 385 = 2640 이 됩니다.

그러면 1부터 100까지 자연수에 대해 "합의 제곱"과 "제곱의 합"의 차이는 얼마입니까?

```python
def getPowerSum (n): ## 제곱의 합
    if n == 1:
        return 1
    else :
        return n**2 + getPowerSum(n-1)
    
def getSumPower (n): ## 합의 제곱
    result = 0
    for x in xrange(n+1):
        result += x
    return pow(result, 2)

print "1부터 10까지 자연수에 대해 제곱의 합은 %d, 합의 제곱은 %d이며, 차는 %d이다." % (getPowerSum(10), getSumPower(10), getSumPower(10) - getPowerSum(10))
print "1부터 20까지 자연수에 대해 제곱의 합은 %d이며, 합의 제곱은 %d, 차는 %d이다." % (getPowerSum(20), getSumPower(20), getSumPower(20) - getPowerSum(20))
```

합의 제곱과 제곱의 합을 구하는 함수를 통해 값을 출력하였다. 합의 제곱은 반복문을 이용해 총 합을 구한 후 pow를 통해 제곱하였다. 제곱의 합은 재귀함수를 사용하였다.
